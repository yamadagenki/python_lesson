```python
import pandas as pd
import numpy as np
from sklearn import tree
%matplotlib inline
import seaborn
seaborn.set()
import re
from IPython.display import display


import warnings
warnings.filterwarnings('ignore')

train = pd.read_csv("./train.csv")
test = pd.read_csv("./test.csv")

def lack_table(df): 
    null_val = df.isnull().sum()
    percent = 100 * null_val / len(df)
    kesson_table = pd.concat([null_val, percent], axis=1)
    kesson_table_ren_columns = kesson_table.rename(
    columns = {0 : '欠損数', 1 : '%'})
    return kesson_table_ren_columns


#-------------------Survived/Died by Class -------------------------------------
# def survived_by_column_name(column_name):
#     survived_class = dataset[dataset['Survived']==1][column_name].value_counts()
#     dead_class = dataset[dataset['Survived']==0][column_name].value_counts()
#     df_class = pd.DataFrame([survived_class,dead_class])
#     df_class.index = ['Survived','Died']
#     df_class.plot(kind='bar',stacked=True, figsize=(5,3), title=("Survived/Died by " + column_name))

# survived_by_column_name("Pclass")    
# survived_by_column_name("Sex")

train["Age"] = train["Age"].fillna(train["Age"].median())
train["Embarked"] = train["Embarked"].fillna(train["Embarked"].mode()[0])
train["Sex"][train["Sex"] == "male"] = 0
train["Sex"][train["Sex"] == "female"] = 1
train["Embarked"][train["Embarked"] == "S" ] = 0
train["Embarked"][train["Embarked"] == "C" ] = 1
train["Embarked"][train["Embarked"] == "Q"] = 2
train["Cabin"] = train["Cabin"].fillna("C")
train["Cabin"][train["Cabin"].str.startswith("A")] = "1"
train["Cabin"][train["Cabin"].str.startswith("B")] = "2"
train["Cabin"][train["Cabin"].str.startswith("C")] = "3"
train["Cabin"][train["Cabin"].str.startswith("D")] = "4"
train["Cabin"][train["Cabin"].str.startswith("E")] = "5"
train["Cabin"][train["Cabin"].str.startswith("F")] = "6"
train["Cabin"][train["Cabin"].str.startswith("G")] = "7"
train["Cabin"][train["Cabin"].str.startswith("T")] = "8"
train["Cabin"] = train["Cabin"].replace("1", 0).replace("2", 1).replace("3", 2).replace("4", 3).replace("5", 4).replace("6", 5).replace("7", 6).replace("8", 7)


test["Age"] = test["Age"].fillna(test["Age"].median())
test["Fare"] = test["Fare"].fillna(test["Fare"].median())
test["Embarked"] = test["Embarked"].fillna(test["Embarked"].mode()[0])
test["Sex"][test["Sex"] == "male"] = 0
test["Sex"][test["Sex"] == "female"] = 1
test["Embarked"][test["Embarked"] == "S" ] = 0
test["Embarked"][test["Embarked"] == "C" ] = 1
test["Embarked"][test["Embarked"] == "Q"] = 2
test["Cabin"] = test["Cabin"].fillna("C")
test["Cabin"][test["Cabin"].str.startswith("A")] = "1"
test["Cabin"][test["Cabin"].str.startswith("B")] = "2"
test["Cabin"][test["Cabin"].str.startswith("C")] = "3"
test["Cabin"][test["Cabin"].str.startswith("D")] = "4"
test["Cabin"][test["Cabin"].str.startswith("E")] = "5"
test["Cabin"][test["Cabin"].str.startswith("F")] = "6"
test["Cabin"][test["Cabin"].str.startswith("G")] = "7"
test["Cabin"][test["Cabin"].str.startswith("T")] = "8"
test["Cabin"] = test["Cabin"].replace("1", 0).replace("2", 1).replace("3", 2).replace("4", 3).replace("5", 4).replace("6", 5).replace("7", 6).replace("8", 7)

# 「train」の目的変数と説明変数の値を取得
target = train["Survived"].values
features_one = train[["Pclass", "Sex", "Age", "Fare", "Cabin"]].values
 
# 決定木の作成
my_tree_one = tree.DecisionTreeClassifier()
my_tree_one = my_tree_one.fit(features_one, target)


# 「test」の説明変数の値を取得
test_features = test[["Pclass", "Sex", "Age", "Fare", "Cabin"]].values

# 「test」の説明変数を使って「my_tree_one」のモデルで予測
my_prediction = my_tree_one.predict(test_features)
  
# PassengerIdを取得
PassengerId = np.array(test["PassengerId"]).astype(int)
 
# my_prediction(予測データ）とPassengerIdをデータフレームへ落とし込む
my_solution = pd.DataFrame(my_prediction, PassengerId, columns = ["Survived"])
 
# my_tree_one.csvとして書き出し
my_solution.to_csv("my_tree_one.csv", index_label = ["PassengerId"])
 
```
